```python
# BeautifulSoup 파싱

# pip          파이썬 전용 모듈 설치 프로그램. 의존성 라이브러리. 라이브러리 갱신

import urllib.request as req
url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
values = {
    'stnId' : 109
}
import urllib.parse as pa
values = pa.urlencode( values )  
url = url + '?' + values

data = ''
with req.urlopen( url ) as f :
    data = f.read().decode( 'utf-8' )
    
from bs4 import BeautifulSoup as bs #beautifulsoup모듈 parser를 도와줌.
soup = bs( data, 'html.parser' ) #data : 링크, html형태를. parser 뽑아내겠다!

title = soup.find( 'title' ).string  # 태그 사이에 문자열 
print( title )

cities = soup.find_all( 'city' )  #find_all : 모두 찾고 리스트로 
for city in cities :
    print( city.string )

datas = soup.find_all( 'data' )
for data in datas :
    print( data.parent.city.string )
    print( data.tmx.string, data.tmn.string )
```

```python
# HTML DOM  Document Object Model
html = """
    <html>
        <head>
            <meta charset='utf-8'>
        </head>
        <body>
            <h2> 저녁메뉴 </h2>
            <p id='first'>중식</p>
            <p class='point strong'>일식</p>
            <p id='second'>한식</p>
            <p class='point'>양식</p>
        </body>
    </html>
"""
from bs4 import BeautifulSoup as bs
soup = bs( html, 'html.parser' )
# find() find_all()                    # 태그 사용
h2 = soup.find( 'h2' )
print( '<<', h2.string, '>>' )
ps = soup.find_all( 'p' )
for p in ps :
    print( '-', p.string )

body = h2.parent
# print( body )
children = body.children
# for child in children :
#     print( child )

h2 = body.findChild()                    # firstChild    
print( h2 )   
p1 = h2.next_sibling.next_sibling
print( p1.string )
p2 = p1.next_sibling.next_sibling
print( p2.string )

# select_one() select()                아이디선택자 클래스선택자 사용
```
