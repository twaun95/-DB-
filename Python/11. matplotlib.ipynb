```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

## Plot ##
s = pd.Series( np.random.randn( 10 ), np.arange( 0, 100, 10 ) )
s = s.cumsum()
# print( s )
plt.plot( s )
plt.show()
```

```python
df = pd.DataFrame( np.random.randn( 10, 4 ), columns=['A', 'B', 'C', 'D'], index=np.arange( 0, 100, 10 ) )
# print( df )
# plt.plot( df )

# plt.plot( df['A'] )
# df['A'].plot()

plt.legend( df.columns )
plt.show()
```

```python
## BarPlot ##
s = pd.Series( np.random.randn( 16 ), index=list('abcdefghijklmnop') )
# s.plot( kind='bar' )
# s.plot( kind='barh' )

plt.bar( s.index, height=s.values )
plt.show()
```

```python
df = pd.DataFrame( np.random.randn( 10, 4 ), columns=['A', 'B', 'C', 'D'], index=list('abcdefghij') )
# print( df )
# df.plot( kind='bar' )

df.loc[:, 'A':'B'].plot( kind='bar', stacked=True )

plt.show()
```

```python
## Histogram ##
#  .   X     Y   
s = pd.Series( np.random.randn( 200 ) )
# s.hist()
s.hist( bins=50 )

plt.show()
```

```python
df = pd.DataFrame( np.random.randn( 200, 4 ), columns=['A', 'B', 'C', 'D'] )
# df.hist( bins=50 )
df.loc[:, ['A','D']].hist( bins=20 )
plt.show()
```

```python
# Scatter Plot ##
#     .
x = np.random.normal( 1, 1, size=( 100, 1) )        # , , 
y = np.random.normal( -2, 4, size=( 100, 1) )        # , , 
# print( x )
xy = np.concatenate( (x,y), axis=1 )
df = pd.DataFrame( xy, columns=['X', 'Y'] )          
#    .   -  - Y,  -  - X

# df.scatter()                        # 
# plt.scatter( x, y )                   # Numpy 
plt.scatter( df['X'], df['Y'] )
plt.xlim( -10, 10 )
plt.ylim( -10, 10 )
plt.xlabel( df.columns[0] )
plt.ylabel( df.columns[1] )
plt.show()
```

```python
##   ##
fig = plt.figure()
fig.size = ( 300, 100 )
fig.subplot = ( 1, 2 )
ax1 = fig.add_subplot( 1, 2, 1 )
s1 = pd.Series( np.random.rand( 50 ), index=np.arange(0, 50) )
ax1.plot( s1 )

ax2 = fig.add_subplot( 1, 2, 2 )
s1.hist( bins=10 )

plt.show()
```

```python
fig = plt.figure()
fig.size = ( 200, 200 )
fig.subplot = ( 1, 1 )
ax = fig.add_subplot( 1, 1, 1 )
x = np.random.normal( 1, 2, size=(200, 1) )
y = np.random.normal( 1, 4, size=(200, 1) )
xy = np.concatenate( (x,y), axis=1 )
# print( xy )
df = pd.DataFrame( xy, columns=['X', 'Y'] )
ax.scatter( x=df['X'], y=df['Y'], color='r', alpha=0.5  )
ax.set_xlim( -10, 10 )
ax.set_ylim( -10, 10 )
ax.set_xlabel( df.columns[0] )
ax.set_ylabel( df.columns[1] )
ax.set_title( 'Distance' )
ax.set_xticks( np.arange( x.min(), x.max(), 2 ) )
ax.set_yticks( np.arange( y.min(), y.max(), 2 ) )



plt.show()
```
